```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("C:/Users/hp/Downloads/Zomato-data-.csv")
print(dataframe.head())
```

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
print(dataframe.isnull().sum())
```

```python
#Exploring Restaurant types
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

Conclusion: The majority of restuarant fall into dining category

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel('Type of Restaurant')
plt.ylabel('Votes')
```

Dining restaurant are preferred by a larger number of individuals

```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes,'name']

print('Restaurant with the maximum votes:')
print(restaurant_with_max_votes)
```

```python
sns.countplot(x=dataframe['online_order'])
```

This suggest that a majority of the restaurant do not accept online orders.

```python
plt.hist(dataframe['rate'],bins=5)
plt.title('Ratings Distribution')
plt.show()
```

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

Majority of couples prefer restaurant with an approximate cost of 300 rupees

```python
plt.figure(figsize=(6,6))
sns.boxplot(x= 'online_order', y = 'rate', data = dataframe)
```

Offline orders received lower ratings in comparison to online orders which obtained excellent ratings

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc='size', fill_value = 0)
sns.heatmap(pivot_table, annot=True,cmap='YlGnBu', fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed in (Type)')
plt.show()
```

With this we can say that dining restaurants primarily accept offline orders whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurants but prefer online ordering at cafes
